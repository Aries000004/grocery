## 拼接

-  pd.concat() 拼接 DataFrame 元素, 会产生重复的索引， 
- 需要重置索引， ignore_index 参数 设置为 True , 忽略原索引，建立新的索引
- sort=False
- DataFrame 拼接方式默认是 外连outer,（并集，没有的会NaN） inner 表示内连， 交集
- join_axes 表示连接轴，参数是一个列表，可以 pd.Index([""]) 表示保留的那些columns 或者 index ,要看 axis 的设定

- pd.concat()  
- axis 拼接的轴, 0 行， 1 列
- join outer 并集， inner 交集
- join_axes 直接指定那些列要进行拼接
- ignore_index 默认值是False 忽略原来的索引，创建新的索引
- keys 传入一个表， 将不同的DataFrme 进行分组， 相当于多加一层索引

```python
df2 = df.copy()
df2
```

```python
df3 = pd.concat((df, df2), axis=0, ignore_index=True, join='inner')
df3
```

```python
df3.loc[0]
```

```python
df3 = pd.concat((df, df2),keys=["one", "two"], join='inner')
df3
```

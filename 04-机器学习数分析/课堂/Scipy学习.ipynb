### 积分运算

- `form scipy.integrate import quad`


```python
n = np.linspace(-np.pi, np.pi, 10000)
```

```python
x = np.sin(n)
y = np.cos(n)
plt.figure(figsize=(4, 4))
plt.plot(x, y, "r")
```

```python
f = lambda x: (1 - x ** 2) ** 0.5

# 边界
b = 1
a = -1
```

```python
# 传入， 积分的函数， 积分的上下边界
quad(f, a, b)
```

```python
x = np.linspace(-1, 1, 100)
y = f(x)
# plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.plot(x, -y)
plt.axis("equal")
```

```python
moon = plt.imread("./imges/moonlanding.png")
```

```python
plt.imshow(moon, cmap="gray")
```

```python
n = moon.mean()
```

```python
from scipy.signal import wiener
```

```python
moon_ = ndimage.gaussian_filter(moon, sigma=n*3)

moon_ = ndimage.median_filter(moon, size=6)

# 维纳滤波
moon_ = wiener(moon, mysize=6)

plt.imshow(moon_, cmap="gray")
```

### 积分

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame

from scipy.integrate import quad
```

### 积分运算

- `form scipy.integrate import quad`


```python
n = np.linspace(-np.pi, np.pi, 10000)
```

```python
x = np.sin(n)
y = np.cos(n)
plt.figure(figsize=(4, 4))
plt.plot(x, y, "r")
```

```python
f = lambda x: (1 - x ** 2) ** 0.5

# 边界
b = 1
a = -1
```

```python
# 传入， 积分的函数， 积分的上下边界
quad(f, a, b)
```

```python
x = np.linspace(-1, 1, 100)
y = f(x)
plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.plot(x, -y)
```

### 异常检查

1. 确定过滤条件

2. 使用过滤条件筛选符合条件的数据

3. 获取符合条件的索引标签

4. 使用drop 函数，通过索引标签删除异常值

```python
data = DataFrame(np.random.randn(1000, 3))
data.head()
```

#### 判断数据是否大于三倍的标准差

```python
condition = np.abs(data) > 3 * data.std()
# 获取包含空值的行
con = condition.any(axis=1)
drop_index = data[con].index
data.drop(drop_index)
```

### map 和 apply

- map 不支持，迭代函数， 因为每次传入的是一个值
- apply 支持迭代函数，因为是将整行或者整列传入到函数

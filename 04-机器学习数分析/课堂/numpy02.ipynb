### 部分排序  np.partition()

- 关于 K 参数， 为 正数时， 将最小的 k 个数放在前面
- K 为 负数时 将最大的 K 个数放在 末尾

```python
ndarray = np.random.randint(0, 100, size=15)
```

```python
ndarray
```

```python
np.partition(ndarray, 5)
```

```python
nd = ndarray
np.partition(nd, -5)
```

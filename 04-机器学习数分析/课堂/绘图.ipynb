```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import Series, DataFrame
```

```python
data1 = Series(data=np.random.randint(1000, 10000, 10),
              index=np.arange(1, 11, 1))
data2 = Series(data=np.random.randint(2000, 20000, 10),
              index=np.arange(1, 11, 1))
data3 = Series(data=np.random.randint(3000, 30000, 10),
              index=np.arange(1, 11, 1))
data4 = Series(data=np.random.randint(4000, 40000, 10),
              index=np.arange(1, 11, 1))
```

```python
price = DataFrame({
    "one": data1,
    "two": data2,
    "three":data3,
    "four": data4
})

price.plot(kind="bar")
```

```python
plt.plot(data1.index, data1.values)
plt.plot(data2.index, data2.values)
plt.legend(["one", "two"])
```

```python
plt.figure(figsize=(12,6))
axes1 = plt.subplot(121)
axes1.plot(data2.index, data2.values,'r', label="one")
axes1.grid(True)

axes2 = plt.subplot(122)
axes2.plot(data2.index, data2.values,'g', label="two")
axes2.grid(True)

plt.legend()
# plt.grid()
```

```python
np.random.seed(0)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(1000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor="g")

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 2, 1, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01)

#ax.w_zaxis.set_major_locator(LinearLocator(10))
#ax.w_zaxis.set_major_formatter(FormatStrFormatter('%.03f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

from mpl_toolkits.mplot3d.axes3d import get_test_data
ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y, Z = get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```

```python

plt.figure(figsize=(6, 10))

axes1 = plt.subplot(211)
axes1.scatter(degs, humi)
axes1.set_title("degs-humi", color="r")

axes2 = plt.subplot(212)
axes2.scatter(degs, speeds)
axes2.set_title("degs-speeds", color="g")
```

```python

plt.axes(polar=True)

index = np.linspace(0, 2*np.pi, 8, endpoint=False)

ave_speed = []
for angle in np.arange(0, 360, 45):
    condition = (degs < (angle + 45)) & (degs > angle)
    ave_speed.append(speeds[condition].mean())

color = np.random.random(size=(8,3))
plt.bar(x=index, height=ave_speed, width=0.8, color=color)
    
```

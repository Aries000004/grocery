## 线性回归

```python
import numpy as np
import pandas as pd

import mglearn
import matplotlib.pyplot as plt
```

```python
mglearn.plots.plot_linear_regression_wave()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_wave(n_samples=60)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)

display(lr.coef_, lr.intercept_)

lr.score(X_train, y_train)

lr.score(X_test, y_test)
```

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y)

lr = LinearRegression().fit(X_train ,y_train)
lr.score(X_train, y_train)

lr.score(X_test, y_test)
```

### 岭回归

```python
from sklearn.linear_model import Ridge

rigde = Ridge().fit(X_train, y_train)

rigde.score(X_train, y_train)

rigde.score(X_test, y_test)
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
ridge10.score(X_test, y_test)
ridge10.score(X_train, y_train)
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
```

```python
r1 = rigde.coef_
r10 = ridge10.coef_
r01 = ridge01.coef_
```

```python
plt.figure(figsize=(12,9))
plt.plot(r1, "sr", label="alpha=1")
plt.plot(r10, "^g", label="alpha=10")
plt.plot(r01, "vy", label="alpha=0.1")

plt.plot(lr.coef_, "o", label="linearRegistession")

plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)

plt.legend()
```

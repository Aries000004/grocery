### pandas 自带绘图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import Series, DataFrame
```

```python
nd = np.linspace(0, 100, num=50)

s = Series(nd)

s.cumsum().plot()
```

### 正态分布

```python
nd1 = np.random.normal(165, 10, size=100)
nd2 = np.random.normal(65, 10, size=100)

# 使用级联连接
nd = np.concatenate([nd1, nd2])
s = Series(nd)
s.hist(bins=100, density=True)
s.plot(kind="kde")  # Destiny
```

```python
nd = np.random.randint(0, 50, size=(100, 5))
df = DataFrame(nd,columns=list("XYABC"))
df.plot(x="X", y="Y", kind="scatter")
```

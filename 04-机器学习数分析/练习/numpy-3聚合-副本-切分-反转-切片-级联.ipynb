## 级联

将两个数组连接在一起
级联的参数是列表, 要加中括号或者小括号

```python
n1 = np.random.randint(0,10,(5,5))
```

```python
n1
```

```python
np.concatenate((n1, n1), axis = 1)  # 简单的将数据连接在一起
```

## 切分

- np.split
- np.vsplit
- np.hsplit

```python
n5 = np.random.randint(0,150,(5,7))
```

```python
n5
```

```python
np.split(n5, (1, 3))
# 在 1  和 3  之前进行切分
```

```python
cat0 = np.split(cat, (20, 40))[1]
```

```python
plt.imshow(cat0)
plt.show()
```

```python
n5
```

```python
np.split(n5, (1,4), axis=1)  # 对列进行切割
```

```python
np.hsplit(n5, (1,4))  # 和axis =1 效果相同
```

```python
np.vsplit(n5, (1,4))
```

## 回归

```python
from sklearn.svm import SVR
```

```python
x = np.random.rand(400, 1) * 15
x_train = np.sort(x, axis=0)
```

```python
y_train = np.sin(x_train) 
```

```python
# 加噪声
y_train[::5] += np.random.randn(80,1)
```

```python
plt.scatter(x_train, y_train)
plt.ylim(-2.5, 2.5)
```

```python
svr_linear = SVR(kernel="linear")
svr_rbf = SVR(kernel="rbf")
svr_poly = SVR(kernel="poly")
```

```python
# 数据预测
x_test = np.arange(0,50, 0.01)[:,np.newaxis]
linear_y = svr_linear.fit(x_train, y_train).predict(x_test)

rbf_y = svr_rbf.fit(x_train, y_train).predict(x_test)

poly_y = svr_poly.fit(x_train, y_train).predict(x_test)
```

```python

plt.figure(figsize=(10,8))
plt.scatter(x_train, y_train)

plt.plot(x_test, linear_y, 'r', label="linear")

plt.plot(x_test, rbf_y, 'c', label="rbf")

plt.plot(x_test, poly_y, 'purple',label="poly")

plt.legend()
```

### matplotlib.pyplot 简单使用
1. title  图标的标题
2. xlabel  X轴的标签和字体大小(fontsize=10)
3. ticket_params  刻度标记的文字的大小
4. plot()  绘制图形, 折线图,  也可以通过改变线条的形状转化为散点图, 效果和 scatter() 相同
5.   plt.axis([0, 12, 0, 12]) # 调整 x y 轴的坐标
6. scatter() 散点图  也可以加参数  scatter(x_values, y_valuse, 'xr')  点的样式 X 颜色 r red
7. show()  显示图形
8. %matplotlib.pyplot inline  在当前打印图像, 没有 inline 会弹出 TKAgg框

```python
import matplotlib.pyplot as plt

%matplotlib inline

def main():
    x_values = [x for x in range(1, 11)]
    y_values = [y for y in range(1, 11)]
    
    plt.title('Square Numner')
    plt.xlabel('value', fontsize=18)
    plt.ylabel('Square', fontsize=18)
    
    plt.tick_params(axis='both', labelsize=14)
    
    plt.scatter(x_values, y_values)
    plt.axis([0, 12, 0, 12]) # 调整位置坐标
    plt.show()



if __name__ == '__main__':
    main()
```

### 绘制正弦曲线

**numpy.linespace()**
    - 指定样本的采样范围和样本的数量
    - sin()  正弦函数


```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
# %numpy inline


def main():
    x_values = np.linspace(0, 2 * np.pi, 1000)
    y_values = np.sin(x_values)
    
    plt.plot(x_values, y_values)
    plt.show()



if __name__ == '__main__':
    main()
```

### 在同一坐标绘制多个图像

```python
# coding: utf-8
import matplotlib.pyplot as plt
import numpy as np

%matplotlib

def main():
    x_values = np.linspace(0, 2 * np.pi, 1000)
    plt.plot(x_values, np.sin(x_values), '--b')
    plt.plot(x_values, np.cos(x_values), '--r')
    plt.show()


if __name__ == '__main__':
    main()
    
    
```

### 在量两个坐标绘制两条曲线

matplotlib.pyplot
- subplot(2, 1, 2) subplot(2, 1, 1)  行数, 列数,  位于第二行
- 在每个行上绘图, 最后同一展示   

```python
# coding: utf-8
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def main():
    x_values = np.linspace(0, 2 * np.pi, 50)
    
    
    plt.subplot(2, 1, 1)
    plt.plot(x_values, np.sin(x_values), 'o-b')
    
    plt.subplot(2, 1, 2)
    plt.plot(x_values, np.cos(x_values), '.-r')
    plt.show()

if __name__ == '__main__':
    main()
    
```

### 绘制直方图
- matplotlib.pyplot
- hist()  

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


def main():
    # 通过random 模块的 normal 函数产生1000 个正态分布的样本
    data = np.random.normal(10.0, 5.0, 1000)
    plt.hist(data, 10)
    plt.show()


if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(1, 3, 50)
y = x ** 2 + 1
plt.plot(x, y)
# x y 轴刻度文字的大小
plt.tick_params(axis='both', labelsize=14)
plt.xlabel('x', fontsize='20')
plt.ylabel('y', fontsize='20')
# x y 轴的坐标范围
plt.axis([0, 4, 0, 11])
plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


# 设置全局 坐标轴字体
mpl.rcParams['xtick.labelsize']=15
mpl.rcParams['ytick.labelsize']=15

np.random.seed(42)

x = np.linspace(0, 5, 100) # 等距离取值
# print(x[1])

# 在数据上添加噪声数据, 
y = 2 * np.sin(x) + 0.3 * x ** 2
y_data = y + np.random.normal(scale=0.3, size=100)

plt.figure('data')
plt.plot(x, y_data, '.')

plt.figure('model')
plt.plot(x, y)

plt.figure('data & model')
plt.plot(x, y, 'k', lw=3)
plt.scatter(x, y_data)

plt.show()
```
